```python
import pandas as pd
import plotly.graph_objs as go
from datetime import datetime
```

```python
def getdata(file_list):
  
  d = {}
  h = {}
  t = {}
  for j in file_list:
    data = pd.read_json(j)
    l = len(data)
    for i in range(l):
      if type(data.messages[i]['payload']) == str:
        d[data.messages[i]['timestamp']] = data.messages[i]['payload']
      else:
        h[data.messages[i]['timestamp']] = data.messages[i]['payload']['humidity']
        t[data.messages[i]['timestamp']] = data.messages[i]['payload']['temperature']

  return d,h,t
```

```python
file_list = ['subscription (2).json','subscription (3).json',
             'subscription (4).json','subscription (5).json',
             'subscription (6).json','subscription (7).json',
             'subscription (8).json','subscription (9).json',
             'subscription (10).json']
```

```python
data, humidity, temperature = getdata(file_list)
```

```python
humidity = pd.DataFrame(humidity, index=['humidity'])
temperature = pd.DataFrame(temperature, index=['temperature'])
data = pd.DataFrame(data, index=['data'])
```

```python
humidity = humidity.transpose()
temperature = temperature.transpose()
data = data.transpose()
```

```python
d = []
for i in humidity.index.tolist():
    d.append(datetime.fromtimestamp(i//1000).date())
    
t = []
for i in humidity.index.tolist():
    t.append(datetime.fromtimestamp(i//1000).time())
    
    
humidity['date'] = d
humidity['time'] = t


d = []
for i in data.index.tolist():
    d.append(datetime.fromtimestamp(i//1000).date())
    
t = []
for i in data.index.tolist():
    t.append(datetime.fromtimestamp(i//1000).time())
    
    
data['date'] = d
data['time'] = t




d = []
for i in temperature.index.tolist():
    d.append(datetime.fromtimestamp(i//1000).date())
    
t = []
for i in temperature.index.tolist():
    t.append(datetime.fromtimestamp(i//1000).time())
    
    
temperature['date'] = d
temperature['time'] = t
```

```python

df = temperature
if isinstance(df, (pd.DatetimeIndex, pd.MultiIndex)):
    df = df.to_frame(index=False)

df = df.reset_index().drop('index', axis=1, errors='ignore')
df.columns = [str(c) for c in df.columns]

chart_data = pd.concat([
    df['date'],
    df['temperature'],
], axis=1)
chart_data = chart_data.sort_values(['date'])
chart_data = chart_data.rename(columns={'date': 'x'})
chart_data = chart_data.dropna()

charts = []
line_cfg = {'line': {'shape': 'spline', 'smoothing': 0.3}, 'mode': 'lines'}
charts.append(go.Scatter(
    x=chart_data['x'], y=chart_data['temperature'], name='temperature', **line_cfg
))
figure = go.Figure(data=charts, layout=go.Layout({
    'legend': {'orientation': 'h'},
    'title': {'text': 'temperature by date'},
    'xaxis': {'title': {'text': 'date'}},
    'yaxis': {'title': {'text': 'temperature'}}
}))



from plotly.offline import iplot, init_notebook_mode

init_notebook_mode(connected=True)
figure.pop('id', None)
iplot(figure)
```

### Important observations

* Temperature and Humidity data are collected for 3-10 minutes for 3 days.
* Pump state data is collected for similar span of time for 5 days.
* Every data is collected with time a gap of 5 seconds.
* The pump was switched on only on first day. That is 19th Oct.
* 68% of temperature and humidity data is collected on 20th of Oct.
* On 20th of Oct Temperature and Humidity were observed in clusters or groups.
* On 19th Data is collected in the evening at 5 pm.
* On 21st its at in the afternoon 12pm.
* On 20th its at 11 am, 11:30 am, 2 pm and 3 pm, 4pm
* As the data on 20th is collected at different intervals of time we can see higher variation in that day's data.
